```
%%writefile ./app.py
from flask_ngrok import run_with_ngrok
from flask import Flask
from flask import make_response

app = Flask(__name__)
UPLOAD_FOLDER = 'static/uploads'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
app.secret_key = "secret key"
run_with_ngrok(app)   #starts ngrok when the app is run

import os
import urllib.request
from flask import Flask, flash, request, redirect, url_for, render_template
from werkzeug.utils import secure_filename
from shutil import copyfile as cp
from temp1 import process_image, custom_colorize
from flask import session
import os.path, os

ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg'])


def allowed_file(filename):
	return '.' in filename and filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS
		
@app.route('/')
def upload_form(): 
	if os.path.isfile('./static/uploads/test_image.jpg'):
		os.remove('./static/uploads/test_image.jpg')
	if os.path.isfile('./static/uploads/result.jpg'):
		os.remove('./static/uploads/result.jpg')
	if os.path.isfile('./static/uploads/result_African.jpg'):
		os.remove('./static/uploads/result_African.jpg')
	if os.path.isfile('./static/uploads/result_Asian.jpg'):
		os.remove('./static/uploads/result_Asian.jpg')
	if os.path.isfile('./static/uploads/result_European.jpg'):
		os.remove('./static/uploads/result_European.jpg')
	
	
	
	return render_template('upload.html')

@app.route('/upload', methods=['POST'])
def upload_image():
	if 'file' not in request.files:
		flash('No file part')
		return redirect(request.url)
	file = request.files['file']
	if file.filename == '':
		flash('No image selected for uploading')
		return redirect(request.url)
	if file and allowed_file(file.filename):
		filename = secure_filename(file.filename)
		filename = "test_image.jpg"
		file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
		# flash('Image successfully uploaded and displayed')
		return render_template('upload.html', filename=filename)
	else:
		flash('Allowed image types are -> png, jpg, jpeg')
		return redirect(request.url)
	
# @app.route('/color/<option>')
# def color(option):
# 	filename = custom_colorize(option)
# 	return redirect(url_for('static', filename='uploads/' + filename), code=301)
 
@app.route('/display/<filename>')
def display_image(filename):
	return redirect(url_for('static', filename='uploads/' + filename), code=301)

@app.route('/custom', methods=['GET'])
def custom():
	cp("/content/Passport-posed-color-from-random-posed-black-n-white-image/static/uploads/test_image.jpg", "/content/Passport-posed-color-from-random-posed-black-n-white-image/Rotate-and-Render/3ddfa/example/Images/test_image.jpg")
	process_image()
	filename = "result.jpg"
	return redirect(url_for('static', filename='uploads/' + filename), code=301)
 
app.run()
```
